```python
import pandas as pd

import sys
sys.path.append("..")
from my_modules import my_functions as mybib

%load_ext autoreload
%autoreload 2
```

```python
with open("../data/tableau/accident_data.csv", "rb") as file:
    accident = pd.read_csv(file)
```

```python
accident = accident.drop("strasse", axis=1)
```

```python
display(accident.head(10).T)
display(accident.dtypes)
```

```python
accident['PLR_ID'] = accident['PLR_ID'].astype(str) # change float to int64
accident['PGR_ID'] = accident['PGR_ID'].astype(str) # change float to int64
```

```python
# accident.columns
```

```python
# rename column names

accident.columns = ['key', 'objectid', 'district_id', 'lor_ab_2021', 'year', 'month', 'hour',
       'weekday', 'accident_cat', 'accident_type1', 'accident_type2', 'light', 'is_bicycle',
       'is_car', 'is_pedestrian', 'is_motorcycle', 'is_truck', 'is_puplic_transport', 'street_condition',
       'longitude', 'latitude', 'PLR_ID', 'PLR_NAME', 'BZR_ID', 'BZR_NAME',
       'PGR_ID', 'PGR_NAME', 'geometry', 'nearest_street']
```

```python
plot_df = accident[['year', 'month', 'hour', 'weekday', 'accident_cat', 'accident_type1', 'accident_type2', 'light',
       'is_bicycle', 'is_car', 'is_pedestrian', 'is_motorcycle', 'is_truck',
       'is_puplic_transport', 'street_condition']]
```

```python
mybib.plot_discrete_var(plot_df)
```

```python
accident['accident_cat'] = accident['accident_cat'].replace({1:'deadly', 2:'seriously_injured', 3:'light_injured' }) #rename accident categories
accident['light'] = accident['light'].replace({0:'daylight', 1:'twilight', 2:'darkness'}) # rename light 
accident['street_condition'] = accident['street_condition'].replace({0:'dry', 1:'wet', 2:'icy'}) # rename  street condition
accident['accident_type2'] = accident['accident_type2'].replace({
        1: 'Driving',
        2: 'Turning/Crossing',
        3: 'Turning/Crossing',
        4: 'Crosswalk',
        5: 'Stationary Traffic',
        6: 'Same Direction',
        7: 'Other'
}) # summarize and rename accident types

accident['weekday'] = accident['weekday'].replace({
        1: 'Sunday',
        2: 'Monday',
        3: 'Tuesday',
        4: 'Wednesday',
        5: 'Thursday',
        6: 'Friday',
        7: 'Saturday'
}) # summarize and rename accident types

accident['date'] = pd.to_datetime(accident['year'].astype(str) + '-' + accident['month'].astype(str) + '-01') # date format
accident['involved'] = accident.apply(mybib.involved_partners, axis=1) # summarize involved vehicles
```

```python
plot_df = accident[['year', 'date', 'weekday', 'accident_cat', 'accident_type2', 'light',
       'involved', 'street_condition']]
plot_df = plot_df.sort_values(by='date')
```

```python

mybib.plot_discrete_var(plot_df)
```
